```python
import requests
import pandas as pd
import datetime
import seaborn as sns
import time
```

```python
# Converte o Tempo
def timestamp_to_date(timestamp): 
    return datetime.utcfromtimestamp(timestamp / 1000).strftime('%Y-%m-%d')

# Pega os 90 dias da moeda e converte em tabela
def getDataFrame(coinId): 
    coinData = f"https://api.coingecko.com/api/v3/coins/{coinId}/market_chart?vs_currency=usd&days=90"
    response = requests.get(coinData)
    coins = response.json()


    closing_prices = {}
    for timestamp,close_price in coins["prices"]:
        date = timestamp_to_date(timestamp)
        closing_prices[date] = close_price  # Vai substituir o preço do mesmo dia, garantindo que seja o último

    df = pd.DataFrame({"Datetime":closing_prices.keys(),coinId: closing_prices.values()})
    df.set_index("Datetime", inplace=True)
    return df

# Pega as 'n' melhores moedas de cripto e junta numa tabela
def getTopCripto(data, n):
    lenData = len(data)
    listCriptoNames = [i["id"] for i in data]
    df = getDataFrame(listCriptoNames[0])
    i=1

    while i < n:
        name = listCriptoNames[i]
        print(name)
        
        try:
            dataCripto = getDataFrame(name)
            df = df.join(dataCripto)
            i+=1
        except KeyError:
            time.sleep(60)
    return df

```

```python
n = 150


url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page={n}&page=1&sparkline=false"
response = requests.get(url)
data = response.json()

df = getTopCripto(data,n)
df.head()
```
